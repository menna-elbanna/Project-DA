```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

plt.style.use("default")
```

```python
df = pd.read_pickle("/content/Data/hotel_bookings_cleaned (1).pkl")
```

```python
plt.figure(figsize=(6,4))
df['hotel'].value_counts().plot(kind='bar')
plt.title("Count of Hotel Types")
plt.xlabel("Hotel Type")
plt.ylabel("Count")
plt.show()
```

```python
plt.figure(figsize=(6,4))
plt.hist(df['lead_time'].dropna(), bins=40)
plt.title("Lead Time Distribution")
plt.xlabel("Lead Time (days)")
plt.ylabel("Frequency")
plt.show()
```

```python
plt.figure(figsize=(12,6))
sns.heatmap(df.corr(numeric_only=True), annot=False, cmap="coolwarm")
plt.title("Correlation Heatmap")
plt.show()
```

```python
plt.figure(figsize=(7,5))
sns.boxplot(x='hotel', y='adr', data=df)
plt.title("ADR by Hotel Type")
plt.xlabel("Hotel Type")
plt.ylabel("ADR")
plt.show()
```
