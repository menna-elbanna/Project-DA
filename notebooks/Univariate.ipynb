```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_pickle('/content/data/hotel_bookings_cleaned (1).pkl')
```

```python
cancel_counts = df['is_canceled'].value_counts()
cancellation_rate = df['is_canceled'].mean() * 100
print("Cancellation Rate (%):", cancellation_rate)
```

```python
plt.figure(figsize=(6,4))
cancel_counts.plot(kind='bar')
plt.title('Cancellation Distribution')
plt.xlabel('is_canceled (0=Not Canceled, 1=Canceled)')
plt.ylabel('Count')
plt.tight_layout()
plt.show()
```

```python
reserved_counts = df['reserved_room_type'].value_counts()
print("Most Reserved Room Type:", reserved_counts.idxmax())
```

```python
plt.figure(figsize=(8,4))
reserved_counts.plot(kind='bar')
plt.title('Reserved Room Type Distribution')
plt.xlabel('Room Type')
plt.ylabel('Count')
plt.tight_layout()
plt.show()
```

```python
assigned_counts = df['assigned_room_type'].value_counts()
print("Most Assigned Room Type:", assigned_counts.idxmax())
```

```python
plt.figure(figsize=(8,4))
assigned_counts.plot(kind='bar')
plt.title('Assigned Room Type Distribution')
plt.xlabel('Room Type')
plt.ylabel('Count')
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(6,4))
plt.hist(df['adr'].dropna())
plt.title('ADR Price Distribution (Histogram)')
plt.xlabel('ADR')
plt.ylabel('Frequency')
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(6,4))
plt.boxplot(df['adr'].dropna())
plt.title('ADR Boxplot')
plt.ylabel('ADR')
plt.tight_layout()
plt.show()
```

```python
Q1 = df['adr'].quantile(0.25)
Q3 = df['adr'].quantile(0.75)
IQR = Q3 - Q1
outliers = df[(df['adr'] < Q1 - 1.5 * IQR) | (df['adr'] > Q3 + 1.5 * IQR)]
print("Number of ADR Outliers:", len(outliers))
```

```python
print("ADR Mean:", df['adr'].mean())
print("ADR Median:", df['adr'].median())
```
