```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
from pathlib import Path

data_path = Path("../processed data/hotel_bookings_cleaned.pkl")
df = pd.read_pickle(data_path)
```

```python
plt.figure(figsize=(6,4))
sns.boxplot(data=df, x='is_canceled', y='adr')
plt.title('ADR vs Cancellation')
plt.xlabel('is_canceled (0=Not Canceled, 1=Canceled)')
plt.ylabel('ADR')
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(6,4))
sns.boxplot(data=df, x='is_canceled', y='lead_time')
plt.title('Lead Time vs Cancellation')
plt.xlabel('is_canceled')
plt.ylabel('Lead Time')
plt.tight_layout()
plt.show()
```

```python
numeric_df = df.select_dtypes(include=['int64', 'float64'])
plt.figure(figsize=(10,8))
sns.heatmap(numeric_df.corr(), annot=False, cmap='coolwarm')
plt.title('Correlation Heatmap')
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(8,4))
df.groupby('reserved_room_type')['adr'].mean().sort_values().plot(kind='bar')
plt.title('Average ADR by Reserved Room Type')
plt.xlabel('Room Type')
plt.ylabel('Average ADR')
plt.tight_layout()
plt.show()
```
