```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_pickle('../processed data/hotel_bookings_cleaned.pkl')
```

```python
plt.figure(figsize=(6,4))
sns.boxplot(data=df, x='is_canceled', y='adr')
plt.title('ADR vs Cancellation')
plt.xlabel('is_canceled (0=Not Canceled, 1=Canceled)')
plt.ylabel('ADR')
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(6,4))
sns.boxplot(data=df, x='is_canceled', y='lead_time')
plt.title('Lead Time vs Cancellation')
plt.xlabel('is_canceled')
plt.ylabel('Lead Time')
plt.tight_layout()
plt.show()
```

```python
numeric_df = df.select_dtypes(include=['int64', 'float64'])
plt.figure(figsize=(10,8))
sns.heatmap(numeric_df.corr(), annot=False, cmap='coolwarm')
plt.title('Correlation Heatmap')
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(8,4))
df.groupby('reserved_room_type')['adr'].mean().sort_values().plot(kind='bar')
plt.title('Average ADR by Reserved Room Type')
plt.xlabel('Room Type')
plt.ylabel('Average ADR')
plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

df = pd.read_pickle('/content/hotel_bookings_cleaned (1).pkl')

adr_by_adults = df.groupby('adults')['adr'].mean().reset_index()

plt.figure(figsize=(8,6))
sns.lineplot(x='adults', y='adr', data=adr_by_adults, marker='o')
plt.title("Average Daily Rate (ADR) by Number of Adults")
plt.xlabel("Number of Adults")
plt.ylabel("Average ADR")
plt.show()
```

```python
df['total_nights'] = df['stays_in_weekend_nights'] + df['stays_in_week_nights']

plt.figure(figsize=(8,6))
sns.scatterplot(x='total_nights', y='adr', data=df)
plt.title("Total Nights vs ADR")
plt.xlabel("Total Nights")
plt.ylabel("ADR")
plt.show()


```

```python
plt.figure(figsize=(10,6))
sns.barplot(x='arrival_date_month', y='adr', data=df, estimator='mean')
plt.title("Average ADR by Arrival Month")
plt.xlabel("Month")
plt.ylabel("Average ADR")
plt.xticks(rotation=45)
plt.show()

```

```python
plt.figure(figsize=(8,6))
sns.barplot(x='adults', y='adr', data=df, estimator='mean')
plt.title("Average ADR by Number of Adults")
plt.xlabel("Number of Adults")
plt.ylabel("Average Daily Rate (ADR)")
plt.show()
```
